### Data Preprocessing
  - **Load the Dataset**: Import the dataset from a CSV file.
  - **Initial Inspection**: View the first few rows of the dataset and check its dimensions.
  - **Handle Missing Values**: Identify missing values in the dataset.
  - **Convert Data Types**: Ensure numerical columns are of the correct type.
  - **One-Hot Encoding**: Convert categorical variables into a format suitable for machine learning models.

```python
import pandas as pd
df = pd.read_csv('../data/GLEAM_LivestockEmissions.csv')

print(df.head())
print("Dataset dimensions:", df.shape)
print("Missing values:\n", df.isnull().sum())

numerical_columns = [
    'Emission Intensity (kg CO2e per kg protein)', 'Production (kg protein)',
    'Total GHG emissions (kg CO2e)', 'Total CO2 emissions (kg CO2e)',
    'Total CH4 emissions (kg CO2e)', 'Total N2O emissions (kg CO2e)',
    'Feed, CO2 (kg CO2e)', 'Feed, CH4 (kg CO2e)',
    'Feed: fertilizer & crop residues, N2O (kg CO2e)',
    'Feed: applied & deposited manure, N2O (kg CO2e)',
    'LUC: soy & palm, CO2 (kg CO2e)', 'LUC: pasture expansion, CO2 (kg CO2e)',
    'Enteric fermentation, CH4 (kg CO2e)', 'Manure management, CH4 (kg CO2e)',
    'Manure management, N2O (kg CO2e)', 'Direct energy, CO2 (kg CO2e)',
    'Indirect energy, CO2 (kg CO2e)', 'Postfarm, CO2 (kg CO2e)'
]

df[numerical_columns] = df[numerical_columns].apply(pd.to_numeric, errors='coerce')

# One-hot encoding for categorical variables
categorical_columns = ['Region', 'Animal species', 'Production system', 'Commodity']
df = pd.get_dummies(df, columns=categorical_columns)

print("Dataset dimensions after one-hot encoding:", df.shape)
print("Basic statistics for numerical columns after encoding:")
df.describe()
```

### Exploratory Data Analysis (EDA)

```python
# Exploratory Data Analysis (EDA)

import matplotlib.pyplot as plt
import seaborn as sns

# Distribution plots for numerical variables
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(5, 4, i)
    sns.histplot(df[col], kde=True)
    plt.title(col, fontsize=8)  
plt.tight_layout()
plt.show()

# Correlation analysis
corr_matrix = df[numerical_columns].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()

# Relationship with categorical variable
region_columns = [col for col in df.columns if col.startswith('Region_')]
print("Region columns:", region_columns)

if region_columns:
    plt.figure(figsize=(12, 8))
    df_melted = df.melt(id_vars=['Total GHG emissions (kg CO2e)'], value_vars=region_columns, var_name='Region', value_name='Present')
    df_melted = df_melted[df_melted['Present'] == 1]
    sns.boxplot(x='Region', y='Total GHG emissions (kg CO2e)', data=df_melted)
    plt.xlabel('Region', fontsize=12)
    plt.ylabel('Total GHG emissions (kg CO2e)', fontsize=12)
    plt.title('Boxplot of Total GHG Emissions by Region', fontsize=14)
    plt.xticks(rotation=45)
    plt.tick_params(axis='x', labelsize=10) 
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Outlier detection
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(5, 4, i)
    sns.boxplot(y=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

### Analysis and Hypothesis Testing
- **Data Preparation**: Create a subset of the dataset with relevant columns and handle missing values.
- **Correlation Analysis**: Calculate the Pearson correlation coefficient between production and total GHG emissions.
- **Hypothesis Testing**: Test the significance of the correlation using a t-test.
- **Visualization**: Create a scatter plot to visualize the relationship between production and total GHG emissions, colored by region.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from scipy.stats import pearsonr, t

df = pd.read_csv('../data/GLEAM_LivestockEmissions.csv')
df_subset = df[['Animal species', 'Production (kg protein)', 'Total GHG emissions (kg CO2e)']]
df_subset = df_subset.dropna(subset=['Production (kg protein)', 'Total GHG emissions (kg CO2e)'])

scaler = MinMaxScaler()
df_subset[['Production (kg protein)', 'Total GHG emissions (kg CO2e)']] = scaler.fit_transform(df_subset[['Production (kg protein)', 'Total GHG emissions (kg CO2e)']])
standardizer = StandardScaler()
df_subset[['Production (kg protein)', 'Total GHG emissions (kg CO2e)']] = standardizer.fit_transform(df_subset[['Production (kg protein)', 'Total GHG emissions (kg CO2e)']])

# Z-score to remove outliers
z_scores = np.abs(stats.zscore(df_subset[['Production (kg protein)', 'Total GHG emissions (kg CO2e)']]))
df_subset = df_subset[(z_scores < 3).all(axis=1)]

# Pearson correlation coefficient
correlation, _ = pearsonr(df_subset['Production (kg protein)'], df_subset['Total GHG emissions (kg CO2e)'])
print(f"Pearson Correlation Coefficient: {correlation:.4f}")

# hypothesis testing for correlation significance
alpha = 0.05  # significance level
n = len(df_subset)

# t-score and p-value for Pearson correlation
r = correlation
t_score = r * np.sqrt((n - 2) / (1 - r**2))
p_value = 2 * (1 - t.cdf(abs(t_score), df=n-2))

print(f"T-Score: {t_score:.4f}")
print(f"P-Value: {p_value:.4f}")

# Scatter plot of Production vs Total GHG emissions by Animal species
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Production (kg protein)', y='Total GHG emissions (kg CO2e)', data=df_subset, hue='Animal species', alpha=0.7, palette='tab10')
plt.title('Production vs Total GHG Emissions by Animal Species')
plt.xlabel('Production (kg protein)')
plt.ylabel('Total GHG emissions (kg CO2e)')
plt.legend(title='Animal Species')
plt.grid(True)
plt.show()
```
